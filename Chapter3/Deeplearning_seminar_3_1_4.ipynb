## 신경망이 처음으로 할 일 : 예측

3장을 통해 신경망의 세 가지 구성품(데이터, 기계, 예측 결과)이 동작하는 내부 원리에 대해 알아보자.

신경망을 만들려면 input data set과 output data set이 어떤 shape(한 번에 처리하는 데이터 요소 개수)인지 먼저 이해해야한다.  
아래는 input data 1개를 입력하여 가중치 0.1을 통해 하나의 예측 결과를 출력하는 신경망이다. 평균 발가락 개수를 통해 야구팀의 승리 가능성을 예측하고자 한다.

## 예측을 수행하는 신경망

```python
#신경망

weight = 0.1

def neural_network(input, weight):
    prediction = input * weight
    return prediction
```

```python
#입력 데이터 요소 하나 집어넣기

number_of_toes = [8.5, 9.5, 10, 9]
input_data = number_of_toes[0]

pred = neural_network(input_data, weight) #입력에 가중치 곱하기
print(pred) #예측 결과 내놓기
```

## 신경망이 뭔가요?

위에서 입력한 코드가 신경망이고 만든 신경망을 통해 예측까지 진행하였다. 예측 결과는 0.85로 올바르게 나왔다.  

신경망이란, 방금 만든 신경망으로 설명하자면 하나 또는 복수 개로 이루어진 가중치이며, 이 가중치는 예측을 위해 input data와 곱할 수 있다고만 하자.  
input data란, 오늘의 온도, 야구 선수의 타율, 어제의 주가 등 대체로 쉽게 파악할 수 있는 data  
prediction이란, 온도를 고려했을 때 오늘 사람들이 운동복을 입을 확률은 0%입니다 등 주어진 입력에 대해 신경망이 말하는 바  
prediction은 항상 맞는가? 아니다. 하지만 그 오류를 통해 배울 수 있다. 이를테면 prediction이 너무 높으면 신경망은 다음 prediction이 낮아지도록 weight를 조정한다.  
신경망은 어떻게 학습하는 건가? 한 마디로 시행착오를 통해 학습한다. 이전 prediction을 통해 동일한 입력이 들어왔을 때 더 정확하게 예측하도록 weight를 올리거나 낮춰서 조정한다.  

## 신경망이 하는 일이 궁금합니다.

#### 신경망은 입력에 가중치를 곱해서 입력을 확대하거나 축소한다.

위에서 만든 신경망은 아주 단순한 형태로 input data(8.5)를  weight(0.1)에 곱한다. 가중치가 2면 신경망은 입력값을 두 배로, 가중치가 0.01이면 입력을 100으로 나눈다. 이처럼 입력을 크거나 작게 만들 수 있다.

```python
weight = 0.1

def neural_network(input, weight):
    prediction = input * weight
    return prediction
```

신경망은 input 변수를 정보로, weight 변수를 지식으로 받아들여 prediction을 출력한다. 신경망은 input data 안에 있는 정보를 해석하기 위해 가중치에 담긴 지식을 활용한다.

우리가 위에서 사용한 '정보'는 발가락 개수이다. 아, 그리고 우리가 알아둬야할 것은

첫 번째, 신경망은 한 번에 오직 1개의 인스턴스(데이터 요소)에만 접근한다. 하지만 신경망에 복수 입력을 한꺼번에 공급함으로써 신경망이 단기 기억을 갖게 하는 방법을 배울 것이다.  
다른 시각으로 보면 신경망의 가중치 값을 신경망의 input과 prediction 사이의 민감도라고 생각할 수 있다. 가중치가 꽤 크다면 작은 input조차도 큰 prediction을 낳을 수 있다. 마치 볼륨처럼 말이다.

```python
number_of_toes = [8.5, 9.5, 10, 9]
input_data = number_of_toes[0]

pred = neural_network(input_data, weight)
print(pred)
```

여기서 신경망이 실제 하는 일은 number_of_toes 변수에 볼륨(가중치)을 연결하는 것이다. 이론 적으로 볼륨은 평균 발가락 개수가 야구팀의 승리 가능성을 알려줄 수 있다. 물론 이 예측은 제대로 작동하지 않을 가능성이 크다. 그렇기에 여러 정보 조각을 신경망에 제공하여 훨씬 정확한 결정을 할 수 있도록 할 것이다.  
신경망 prediction에는 양수만 있는 것이 아니라 음수도 존재한다. input 또한 마찬가지다. 온도가 -10도일 때 패딩을 입을 확률이 90% 이상이라고 예측할 것이다.
